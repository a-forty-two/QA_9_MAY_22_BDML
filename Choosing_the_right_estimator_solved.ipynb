```python
# ONLY nums!-> diagnosis! (LABEL)
cancer_dict = {'M': 1, 'B':0}
y = df.pop('diagnosis')
df.head()
# y = weights * df + bias
```

```python
# X -> normalized
# Y -> encoded 
```

```python
y_encoded = y.map(lambda label: cancer_dict[label] )
y_encoded.head()
```

```python
# Norm
stats = df.describe().T
stats
```

```python
mu = stats['mean']
sigma = stats['std']
x = (df - mu)/sigma
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y_encoded,test_size=0.2,
                                                    random_state=42)
```

```python
ytrain.head()
```

# Part 3: Model Selection

### Q. Use the cheat sheet below to choose the algorithm/estimator suitable for building a model to address your candidate question(s)

* https://scikit-learn.org/stable/tutorial/machine_learning_map/

```python
import numpy as np
from sklearn.linear_model import SGDClassifier # gradient descent
from sklearn.svm import SVC, LinearSVC # graph
from sklearn.naive_bayes import GaussianNB # probability
from sklearn.tree import DecisionTreeClassifier # TREE based 
from sklearn.ensemble import RandomForestClassifier 
# I need some METRICS from which i could compare these 3 algorithms and select which was better or worse
import time 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
algos = {'KNN':KNeighborsClassifier(),'RFPlain': RandomForestClassifier(),'RF5':RandomForestClassifier(n_estimators=5),'RF200':RandomForestClassifier(n_estimators=200), 'Dtree':DecisionTreeClassifier(),'SGD':SGDClassifier(), 'LinearSVC':LinearSVC(), 'SVC':SVC(), 'Sherlock':GaussianNB()}
# all algos that I want to compare 

time_all = [] 
acc_all = []
cvs_all = []
```

```python
best_algo_acc = 0
best_model = NotImplementedError
```

```python
for algo_name, algo in algos.items():
  start_time = time.time()
  algo.fit(xtrain, ytrain)# training
  p = algo.predict(xtest) # scoring
  acc = accuracy_score(p, ytest) #evaluation
  if acc > best_algo_acc:
    best_algo_acc = acc
    best_model = algo
  cv_score = cross_val_score(algo, xtest, ytest, cv=5)
  end_time = time.time()
  timetaken = end_time - start_time
  time_all.append(timetaken)
  acc_all.append(acc)
  cvs_all.append(cv_score)
  print('**************')
  print(algo_name)
  print("time taken = " + str(timetaken))
  print("Accuracy = " + str(acc))
  print("Cross Val Score = " + str(cv_score))

```
